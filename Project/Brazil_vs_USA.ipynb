```python
df = pd.read_csv('Full Brazil America Dataset.csv')
countries = ['United States of America', 'Brazil']
df=df[df['location'].isin(countries)]
df.head()
```

```python
df= pd.read_csv('Fulldatasetincludingages.csv')
countries = ['United States of America', 'Brazil']
df=df[df['location'].isin(countries)]
df = df.replace('years','',regex=True)
df = df[df['age']!='All ages']
df = df[df['metric']=='Number']
df = df[df['year']==2018]
```

```python
def get_median_age(age_range):
    age_split = age_range.split('-')
    return (int(age_split[0]) + int(age_split[1])) / 2

df['Age'] = df['age'].apply(get_median_age)
df

```

```python
df['Number of Diagnoses']=df['val']
sns.lineplot(data=df,x='Age',y='Number of Diagnoses',hue='location').set_title('Number of Diagnoses of Type 2 Diabetes in 2018')
```

```python
df['Proportion of Diagnoses']=df['val']/(df['val'].sum())
dfB = df[df['location']=='Brazil']
sns.lineplot(data=dfB,x='Age',y='Proportion of Diagnoses', color='green').set_title('Distribution of Diagnoses of Type 2 Diabetes in Brazil in 2018')
dfB.describe()
```

```python
dfU = df[df['location']=='United States of America']
sns.lineplot(data=dfU,x='Age',y='Proportion of Diagnoses',color='orange').set_title('Distribution of DIagnoses of Type 2 Diabetes in the USA in 2018')
dfU.describe()
```
