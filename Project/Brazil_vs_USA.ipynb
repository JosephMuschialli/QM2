```python
import pandas as pd
import matplotlib
import pylab
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('ggplot')
pylab.rcParams['figure.figsize'] = (10., 8.)
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv('Full Brazil America Dataset.csv')
countries = ['United States of America', 'Brazil']
df=df[df['location'].isin(countries)]
df=df[df['metric']=='Number']
df
```

```python
df['Number of Incidences of Diabetes (In Millions)']=df['val']/1000000
df['Year']=df['year']
df['Country']=df['location']
sns.lineplot(data=df,x='Year',y='Number of Incidences of Diabetes (In Millions)',hue='Country').set_title('Number of Incidences of Type 2 Diabetes over Time')
```
