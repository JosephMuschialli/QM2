```python
df = pd.read_csv('Full Brazil America Dataset.csv')
df=df[df['metric']=='Number']
df=df[df['age']=='All ages']
df.head()
```

```python
# This is old code just in case we need it:
# df['Number of Incidences of Diabetes (In Millions)']=df['val']/1000000
# df['Year']=df['year']
# df['Country']=df['location']
# sns.lineplot(data=df,x='Year',y='Number of Incidences of Diabetes (In Millions)',hue='Country').set_title('Number of Incidences of Type 2 Diabetes over Time')

# prepare data set

df['val_in_ht']=df['val']/100000
df_B = df[df['location']=='Brazil']

plt.figure(figsize=(8,6))

# plot line
plt.plot(df_B['year'], df_B['val_in_ht'],
         lw=2.5,
         color='green')
plt.scatter(df_B['year'], df_B['val_in_ht'],
            color='green')

# set label
plt.xlabel('Year')
plt.ylabel('Number of Incidences of Diabetes, in hundred thousands')

plt.title('Number of Incidences of Diabetes Over Time in Brazil')
plt.show()
```

```python
# Now, the same for all states in Brazil
dfsB = pd.read_csv('Brazilnewlocationdata.csv')
dfsB = dfsB[dfsB['measure']=='Incidence']
dfsB=dfsB[dfsB['age']=='All ages']
dfsB = dfsB[dfsB['metric']=='Number']
dfsB['Number of Incidences of Diabetes (In Ten Thousands)']=dfsB['val']/10000
sns.lineplot(data=dfsB,x='year',y='Number of Incidences of Diabetes (In Ten Thousands)',hue='location').set_title('Number of Incidences of Type 2 Diabetes over Time in Brazil')
plt.legend(ncol=1,
           bbox_to_anchor=(1.05, 1), loc='upper left')
```

```python
# Let's replicate this for only the top ten states as of 2018
dfsB = dfsB[dfsB['Number of Incidences of Diabetes (In Ten Thousands)']>=1.5]
sns.lineplot(data=dfsB,x='year',y='Number of Incidences of Diabetes (In Ten Thousands)',hue='location').set_title('Number of Incidences of Type 2 Diabetes over Time in Brazil (12 Highest States)')
plt.legend(ncol=1,
           bbox_to_anchor=(1.05, 1), loc='upper left')
```
