```python
import pandas as pd
import matplotlib
import pylab
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
pylab.rcParams['figure.figsize'] = (10., 8.)
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv('databytime.csv')
df = df[df['metric']=='Number']
print(df.head())
xpoints = df['year']
ypoints = df['val']
plt.plot(xpoints, ypoints)
plt.title('Newly Diagnosed Type 2 Diabetes Cases over Time in Brazil')
plt.xlabel('Year')
plt.ylabel('Number of New Diagnoses of Type 2 Diabetes / Mid-Year Population [x10**9]')
# The value is taken as the number of newly diagnosed cases of type 2 diabetes.
# Expressed as number of new diagnosis divided by the mid-year population size
```
